```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Load the diabetes dataset
diabetes = pd.read_csv('/content/diabetes.csv.csv')

# Take a random sample of 25 observations
sample = diabetes.sample(25)

# 1 Calculate mean and highest glucose values for sample and population
sample_mean_glucose = sample['Glucose'].mean()
sample_highest_glucose = sample['Glucose'].max()
population_mean_glucose = diabetes['Glucose'].mean()
population_highest_glucose = diabetes['Glucose'].max()

# Create a bar plot to compare mean and highest glucose values
labels_glucose = ['Sample Mean Glucose', 'Sample Highest Glucose', 'Population Mean Glucose', 'Population Highest Glucose']
values_glucose = [sample_mean_glucose, sample_highest_glucose, population_mean_glucose, population_highest_glucose]

plt.figure(figsize=(10, 6))
plt.bar(labels_glucose, values_glucose, color=['blue', 'green', 'orange', 'red'])
plt.title('Comparison of Glucose Statistics')
plt.ylabel('Glucose Value')
plt.show()

# Print mean and highest glucose values
print("Sample Mean Glucose:", sample_mean_glucose)
print("Sample Highest Glucose:", sample_highest_glucose)
print("Population Mean Glucose:", population_mean_glucose)
print("Population Highest Glucose:", population_highest_glucose)

#2  Calculate the 98th percentile of BMI for sample and population
sample_98th_percentile_bmi = np.percentile(sample['BMI'], 98)
population_98th_percentile_bmi = np.percentile(diabetes['BMI'], 98)

#  Create a bar plot to compare the 98th percentile of BMI
labels_bmi = ['Sample 98th Percentile BMI', 'Population 98th Percentile BMI']
values_bmi = [sample_98th_percentile_bmi, population_98th_percentile_bmi]

plt.figure(figsize=(8, 5))
plt.bar(labels_bmi, values_bmi, color=['blue', 'orange'])
plt.title('Comparison of 98th Percentile of BMI')
plt.ylabel('BMI')
plt.show()

# 9. Print the 98th percentile of BMI
print("Sample 98th Percentile BMI:", sample_98th_percentile_bmi)
print("Population 98th Percentile BMI:", population_98th_percentile_bmi)


#3
# Define a function for bootstrap resampling
def bootstrap_resample(data, n_samples):
    return [np.random.choice(data, size=len(data), replace=True) for _ in range(n_samples)]

# Perform bootstrap resampling
n_samples = 500
sample_size = 150
bootstrap_samples = bootstrap_resample(diabetes['BloodPressure'], n_samples)

# Calculate statistics for each bootstrap sample
bootstrap_means = [np.mean(sample) for sample in bootstrap_samples]
bootstrap_stds = [np.std(sample) for sample in bootstrap_samples]
bootstrap_percentiles = [np.percentile(sample, 95) for sample in bootstrap_samples]

# Calculate population statistics
population_mean_bp = diabetes['BloodPressure'].mean()
population_std_bp = diabetes['BloodPressure'].std()
population_percentile_bp = np.percentile(diabetes['BloodPressure'], 95)

# Compare statistics with population statistics using visualizations
plt.figure(figsize=(10, 6))

# Histogram of bootstrap sample means
plt.subplot(2, 2, 1)
plt.hist(bootstrap_means, bins=30, color='blue', alpha=0.7)
plt.axvline(population_mean_bp, color='red', linestyle='--', label='Population Mean')
plt.text(0.5, 0.95, f"Population Mean: {population_mean_bp:.2f}", transform=plt.gca().transAxes, ha='center', va='top')
plt.title('Bootstrap Sample Means')
plt.legend()

# Histogram of bootstrap sample standard deviations
plt.subplot(2, 2, 2)
plt.hist(bootstrap_stds, bins=30, color='green', alpha=0.7)
plt.axvline(population_std_bp, color='red', linestyle='--', label='Population Std')
plt.text(0.5, 0.95, f"Population Std: {population_std_bp:.2f}", transform=plt.gca().transAxes, ha='center', va='top')
plt.title('Bootstrap Sample Standard Deviations')
plt.legend()

# Histogram of bootstrap sample 95th percentiles
plt.subplot(2, 2, 3)
plt.hist(bootstrap_percentiles, bins=30, color='orange', alpha=0.7)
plt.axvline(population_percentile_bp, color='red', linestyle='--', label='Population 95th Percentile')
plt.text(0.5, 0.95, f"Population 95th Percentile: {population_percentile_bp:.2f}", transform=plt.gca().transAxes, ha='center', va='top')
plt.title('Bootstrap Sample 95th Percentiles')
plt.legend()

plt.tight_layout()
plt.show()

# Print the calculated statistics
print("Average Mean Blood Pressure (Bootstrap):", np.mean(bootstrap_means))
print("Average Standard Deviation (Bootstrap):", np.mean(bootstrap_stds))
print("Average 95th Percentile (Bootstrap):", np.mean(bootstrap_percentiles))
print("Population Mean Blood Pressure:", population_mean_bp)
print("Population Standard Deviation:", population_std_bp)
print("Population 95th Percentile:", population_percentile_bp)
```
